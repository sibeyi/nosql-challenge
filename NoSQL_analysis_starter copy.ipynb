### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

local_authority_query = {'$match': {'scores.Hygiene': 0}}

local_authority_groups_query = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1 }}}

local_authority_sort = {'$sort': { 'count': -1 }}

# pipeline (query, group, sort)
pipeline = [local_authority_query, local_authority_groups_query, local_authority_sort]
```

```python
# Run the pipeline through the aggregate method and save the results to a variable
local_authority_results = list(establishments.aggregate(pipeline))
```

```python
# Print the number of documents in the result
print(f"establishments had {len(local_authority_results)} documents in pipeline grouped local authority with hygine score 0 \n"
      f"\n")
```

```python
# Print the first 10 results
pprint(local_authority_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
local_authority_result_df = pd.DataFrame(local_authority_results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(local_authority_result_df))

# Display the first 10 rows of the DataFrame
local_authority_result_df.head(10)
```
